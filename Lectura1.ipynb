# Ver la figura

```python
plt.figure(figsize=(6,3),tight_layout=True)
plt.imshow(img_color)
plt.show()
```

# Slicing

Ir visualizando trozos de la imagen


```python
 plt.figure(figsize=(6, 3), tight_layout=True)
 plt.imshow(img_color[100:500, 200:300]);
 plt.show()
```

# Trasposición de imagen

```python
 plt.figure(figsize=(6, 3), tight_layout=True)
 plt.imshow(np.transpose(img_color, axes=[1, 0, 2]));
 plt.show()
```

# Inversión de imagen

```python
plt.figure(figsize=(6, 3), tight_layout=True)
plt.imshow(img_color[::-1, :, :]);
plt.show()
```

# Espejado de imagen

```python
plt.figure(figsize=(6, 3), tight_layout=True)
plt.imshow(img_color[:, ::-1, :]);
plt.show()
```
